```python
import re 

import pandas as pd 
import matplotlib.pyplot as plt 
import datetime 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
```

```python
fixed_df_train = pd.read_csv('train_df.csv', delimiter=',')
```

```python
print(fixed_df_train.shape)
```

```python
# Разделение данных на обучающее и тестовое подмножества
fixed_df_train = fixed_df_train.drop('search_id', axis=1)
X = fixed_df_train.drop('target', axis=1)
y = fixed_df_train['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
fixed_df_train.head(100000)
```

```python
# Масштабирование данных
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Определение параметров
activation = 'relu'
hidden_layers = [100, 50]
loss = 'binary_crossentropy'
optimizer = 'adam'

# Создание модели
model = MLPClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
# Оценка точности
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Точность:", accuracy)
```

```python

df = pd.DataFrame({'feature1': [1, 2, 3], 'feature2': [4, 5, 6], 'target prediction': [7, 8, 9], 'relevance': [3, 1, 2]})

# Отсортируйте элементы по релевантности
df = df.sort_values(by='relevance', ascending=False)

# Инициализация
p = 2
DCG = []

# Расчет DCG
for i in range(df.shape[0]):
    relevance_position = i + 1
    gain = df['relevance'].iloc[i] * (1 / (p ** relevance_position))
    DCG.append(gain)

DCG = sum(DCG)
ideal_DCG = df['relevance'].sum() * (1 - (1 / (p ** (df.shape[0] + 1))))
DCG = DCG / ideal_DCG

print("DCG:", DCG)
```
